Color, Linestyle, Marker

```python
%matplotlib inline
from matplotlib import pyplot as plt

hari = [4,6,10,13,16,20,22,24,26,28,30]
young_lex = [100,120,160,180,210,240,270,300,320,330,350]

# color. untuk svarna garis
# LinestyLe, untuk jenis garis
# -.
# --
# marker, untuk penanda pertemuan sumbu x don y
# p, pentagon
# D, diamond
plt.plot(hari,young_lex, color='#444444', linestyle='--', marker='D')

atta = [120,140,200,230,260,280,290,320,360,380,400]
plt.plot(hari,atta, color='#5A7D9A', marker='.')

plt.title('Kenaikan Subscriber Berdasarkan Hari')
plt.xlabel('Hari')
plt.ylabel('Jumlah Subscriber')

plt.legend(['Young Lex','Atta Halilintar'])

plt.show()
```

Linewidth

```python
%matplotlib inline

from matplotlib import pyplot as plt

hari = [4,6,10,13,16,20,22,24,26,28,30]
young_lex = [100,120,160,180,210,240,270,300,320,330,350]
plt.plot(hari,young_lex, color='#444444', linestyle='--', marker='o')

# linewidth, untuk menebalkan garis
atta = [120,140,200,230,260,280,290,320,360,380,400]
plt.plot(hari,atta, color='#5A7D9A', marker='D', linewidth='3')

plt.title('Kenaikan Subscriber Berdasarkan Hari')
plt.xlabel('Hari')
plt.ylabel('Jumlah Subscriber')

plt.legend(['Young Lex','Atta Halilintar'])

plt.show()
```

Grid

```python
%matplotlib inline

from matplotlib import pyplot as plt

hari = [4,6,10,13,16,20,22,24,26,28,30]
young_lex = [100,120,160,180,210,240,270,300,320,330,350]
plt.plot(hari,young_lex, color='#444444', linestyle='--', marker='o')

atta = [120,140,200,230,260,280,290,320,360,380,400]
plt.plot(hari,atta, color='#5A7D9A', marker='D', linewidth='3')

plt.title('Kenaikan Subscriber Berdasarkan Hari')
plt.xlabel('Hari')
plt.ylabel('Jumlah Subscriber')

plt.legend(['Young Lex','Atta Halilintar'])

# grid, untuk menampilkan layout grid (susunan vertikal dan horizontal)
plt.grid(True)

plt.show()
```

Menggunakan Style (fivethirtyeight) dan Tight Layout

```python
%matplotlib inline

from matplotlib import pyplot as plt

# menggunakan style fivethirtyeight
plt.style.use('fivethirtyeight')
# plt.style.use('ggplot')

hari = [4,6,10,13,16,20,22,24,26,28,30]
young_lex = [100,120,160,180,210,240,270,300,320,330,350]
plt.plot(hari,young_lex, color='#444444', linestyle='--', marker='o')

atta = [120,140,200,230,260,280,290,320,360,380,400]
plt.plot(hari,atta, color='#5A7D9A', marker='D', linewidth='3')

plt.title('Kenaikan Subscriber Berdasarkan Hari')
plt.xlabel('Hari')
plt.ylabel('Jumlah Subscriber')

plt.legend(['Young Lex','Atta Halilintar'])
plt.grid(True)

# tight layout(), digunakan untuk menyesuaikan plot, label, title di dalam area gambar
plt.tight_layout()
plt.show()
```
